```python
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
df.columns
```

```python
df.dtypes
```

```python
df = df[pd.Index([col for col in df.columns if col[-5:] != "_err1" and col[-5:] != "_err2"])]
```

```python
df.head()
```

```python
df["koi_disposition"] = pd.Categorical(df["koi_disposition"],categories=["CANDIDATE","FALSE POSITIVE","NOT DISPOSITIONED","CONFIRMED"],ordered=True)
```

```python
df["koi_disposition"] = df.koi_disposition.cat.codes
```

```python
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
selected_features = df
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
y = selected_features.pop("koi_disposition")
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=29)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a StandardScater model and fit it to the training data
X_scaler = MinMaxScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from sklearn.ensemble import GradientBoostingClassifier as GBC
# learning_rate=0.1, max_depth=5, min_samples_split=0.1, n_estimators=100, n_iter_no_change=20, random_state=29, score=0.912
model1 = GBC(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=29).fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model1.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model1.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
params = {
    'n_iter_no_change': [2,10,20],
    'n_estimators': [10,100,200],
    'learning_rate': [1.0,.1,.01],
    'max_depth': [1,2,3,5],
    'min_samples_split': [.25,.2,.1],
    'random_state': [29]
}
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV as GS

grid1 = GS(model1,params,verbose=3)
```

```python
# Train the model with GridSearch
grid1.fit(X_train_scaled, y_train)
results = grid1.cv_results_
```

```python
print(grid1.best_params_)
print(grid1.best_score_)
```

```python
grid1.score(X_test_scaled,y_test)
```
